## Data Preprocessing(Normalization, Standardization)

```python
def func_minmaxscale (number, col):
    mx = x_data[:,col].max()
    mn = x_data[:,col].min()
    # (number - min) / (max - min)  -> [0, 1]
    return (number-mn)/(mx-mn)

func_minmaxscale(x_data[0,2],2)
```

sklearn document 웹페이지 : https://scikit-learn.org/stable/modules/classes.html
https://scikit-learn.org/stable/modules/classes.html#module-sklearn.preprocessing

```python
from sklearn.preprocessing import MinMaxScaler
x_scaler = MinMaxScaler()

# fit 메서드 : 데이터에 대해서 fit
x_scaler.fit(x_data)
# transform 메서드: 목표로 하는 데이터를 minmax-scaling

x_data_scaled = x_scaler.transform(x_data)

df_minmax = pd.DataFrame(x_data_scaled, columns=housing.feature_names)
df_minmax['y'] = housing.target
df_minmax.head()
```

```python
def func_std(data):
    m = data.mean()
    s = data.std()

    return (data-m)/s

print(x_data_scaled[:5, 0])
print(func_std(x_data_scaled[:5, 0]))
```

## Standard Normal Distribution

```python
# Gaussian
mu = 3
sigma = 2

x = np.arange(-10, 10, 0.1)
y = (1 / np.sqrt(2 * np.pi * sigma**2)) * np.exp(-(x-mu)**2 / (2 * sigma**2))

plt.plot(x,y)
plt.show()
```
